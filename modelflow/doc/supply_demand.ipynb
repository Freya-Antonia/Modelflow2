```python
# Use LaTeX throughout the figure for consistency
# rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 16})
rc('text', usetex=True)
# Figure dpi
dpi = 72

def ds_plot(f, demand, supply):
    """Make a supply demand plot 


    """
    x = np.linspace(0, 2, 500)
    fig = plt.figure(figsize=(600/dpi, 450/dpi), dpi=dpi)
    ax = fig.add_subplot(111)

    # Plot y = f(x) and y = x
    ax.plot(x, f(x,demand), c='red', lw=2)
    ax.plot(x, f(x,supply), c='blue', lw=2)

    # Annotate and tidy the plot.
    ax.minorticks_on()
    ax.grid(which='minor', alpha=0.5)
    ax.grid(which='major', alpha=0.5)
    #ax.set_aspect('equal')
    ax.set_xlabel('$Price$')
    ax.set_ylabel('$Quantity$')
    demandannotate = f'$Demand = {demand[0]}{demand[1]:+.2f} \\times price {demand[2]:+.1f} \\times price^2$'
    supplyannotate = f'$Supply = {supply[0]}{supply[1]:+.2f} \\times price {supply[2]:+.1f} \\times price^2$'
    ax.set_title('Solving Demand and supply')
    ax.annotate(demandannotate, xy=(0.19,f(0.19, demand)), xytext=(0.35, 1.0),
                arrowprops=dict(facecolor='red', shrink=0.05),)
    ax.annotate(supplyannotate, xy=(0.19,f(0.19, supply)),  xytext=(0.35, 0.0),
                arrowprops=dict(facecolor='blue', shrink=0.05),)

    #plt.savefig('cobweb_{:.1}_{:.2}.png'.format(x0, r), dpi=dpi)
    plt.show()

func = lambda x,ds: ds[0]+ds[1]*x + ds[2]*x*x
demand = 1.,-0.4, 0.1
supply = 0, 0.5, 0.0 
ds_plot(func, demand,  supply)
```
